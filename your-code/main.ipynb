```python
#Help link - https://stackoverflow.com/questions/51034307/pandas-mathematical-operation-conditional-on-column-value

#Create a new column "adjusted_gre" with the new score based on the given condition

admission.loc[admission.university_rating >= 4, "adjusted_gre"] = admission.gre_score+10


#Change "adjusted_gre" position to be near "gre_score"

admission_column_rearrange = admission[['serial_no', 'gre_score', 'adjusted_gre', 'toefl_score', 'university_rating', 'sop',
       'lor', 'cgpa', 'research', 'chance_of_admit']]
admission_column_rearrange
```
